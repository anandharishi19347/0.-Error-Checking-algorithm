```python
#importing the Libraries#
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.cluster import AffinityPropagation
from sklearn import metrics
from sklearn.metrics import pairwise_distances
from sklearn.datasets import make_blobs
```

```python
#Reading the Dataset#
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
#what saved in this file#
dataset
```

```python
X = dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
clusmodel = AgglomerativeClustering(n_clusters = 10)
labels = clusmodel.fit_predict(X)
```

```python
# Fit Affinity Propagation model
affinity_propagation = AffinityPropagation(random_state=15)
affinity_propagation.fit(X)
```

```python
labels
```

```python
supervised = pd.DataFrame(dataset)
```

```python
supervised['Cluster_group']= labels
```

```python
supervised
```

```python
# Number of clusters found
n_clusters_ = len(labels)
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5],fit_reg=False,legend=True, legend_out=True)
```
